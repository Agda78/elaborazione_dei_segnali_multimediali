# Primo Esercizio

```python
# Librerie
import numpy as np
import matplotlib.pyplot as plt
import scipy.ndimage as ndi
import skimage.io as io
```

### Media
Definizione della funzione

```python
def medie(x,K):
    y = ndi.generic_filter(x, np.mean, (K,K))
    return y
```

Utilizzo della funzione

```python
possibilities = [3,5,7,9]
x = io.imread("img/dorian.jpg")
(M,N) = x.shape

plt.figure(figsize=(10, 10))  # Imposta una dimensione adeguata per la figura

for idx, i in enumerate(possibilities, 1):
    y = medie(x, i)
    plt.subplot(2, 2, idx)  # Organizza i subplot in una griglia 2x2
    plt.imshow(y, clim=[0, 255], cmap="gray")  # Mostra l'immagine filtrata
    title = f"Media con finestra: {i}x{i}"
    plt.title(title)
    plt.axis("off")  # Rimuove gli assi per una visualizzazione più pulita
```

### Varianza
Definizione della funzione

```python
def varianze(x,k):
    y = ndi.generic_filter(x, np.var, (k,k))
    return y
```

Utilizzo della funzione

```python
x = io.imread("img/filamento.jpg")

y = varianze(x,3)

plt.figure()
plt.subplot(1,2,1)
plt.imshow(x, clim=[0,255], cmap="gray")
plt.subplot(1,2,2)
plt.imshow(y, clim=[0,255], cmap="gray")
```

### ISTOGRAMMA

```python
plt.figure()

plt.subplot(1,2,1)
plt.hist(x)
plt.title("IS Immagine")

plt.subplot(1,2,2)
plt.hist(y)
plt.title("IS Varianza")
```
