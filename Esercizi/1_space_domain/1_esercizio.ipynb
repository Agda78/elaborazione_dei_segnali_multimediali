# PRIMO ESERCIZIO

```python
# Librerie
import numpy as np
import matplotlib.pyplot as plt
import scipy.ndimage as ndi
import skimage.io as io
```

### Scrittura delle funzioni

```python
def vediJPG(nomefile):
    x = io.imread(nomefile)

    plt.figure()
    plt.imshow(x, clim=[0,255], cmap="gray")
```

```python
def vediRaw(nomefile, nRighe, nColonne, dtype):
    x = np.fromfile(nomefile,dtype)
    x = np.reshape(x, (nRighe, nColonne))

    plt.figure()
    plt.imshow(x, clim=[0,255], cmap="gray")
```

Utilizzo delle Funzioni

```python
vediJPG("img/dorian.jpg")
vediRaw("img/house.y", 512, 512, np.uint8)
```

# Secondo Esercizio

```python
# Carico le classiche immagini RGB
R = io.imread("img/Washington_red.tif")
G = io.imread("img/Washington_green.tif")
B = io.imread("img/Washingyon_blue.TIF")

# Carico l'immagine in infrarossi
I = io.imread("img/Washington_infrared.tif")

# Genero l'immagine con sovrapposizione senza infrarossi
x = np.stack((R,G,B), -1)

# Genero immagine con infrarossi
y = np.stack((I,G,B), -1)

# Plot dell'immagine
plt.figure()
plt.imshow(x)

plt.figure()
plt.imshow(y)
```
