### Esercizio 1

```python
# Funzione da costruire
def clip(x, limit):
    x[x > limit] = limit
    x[x < -limit] = -limit
```

```python
# Esecuzione

# Valori in ingresso alla funzione
x = np.array([-10, 3, -6, 0, 1, -2, 3, 4, -15, 3, 21])
limit = 8

# Utilizzo e stampa dei risultati
print(x)
clip(x,limit)
print("Clip executed: ")
print(x)
```

### Esercizio 2

```python
def clip_matrix(x):
    # Linearizzo la matrice e calcolo il suo ordinamento come un vettore
    sorted = np.sort(x, axis=None)

    quarto_minore = sorted[x.size - 4]
    
    # Dato che devo conservare solo i primi 4 valori massimi, allora vado a prendermi il 4 valore più grande
    # in modo che posso annullare easy gli altri
    x[x < quarto_minore] = 0
    
```

```python
limite = 100
size_len = (16,16) # Lunghezza della matrice (16x16 in questo caso)
x = np.random.randint(limite + 1, size=size_len)

print(x)
clip_matrix(x)
print("Cropped:\n")
print(x)
```
